# Visualization Preparation

Pada bagian ini kita akan mempersiapkan visualisasi-visualisasi yang akan kita gunakan untuk membuat dashboard. Kita akan memberikan analisis data pinjaman yang meliputi:

- **Overview**: Provides a summary of key loan metrics.
- **Time-Based Analysis**: Shows trends over time and loan amounts.
- **Loan Performance**: Analyzes loan conditions and distributions.
- **Financial Analysis**: Examines loan amounts and distributions based on conditions.

````python
import pandas as pd
import plotly.express as px

# pandas output display setup
pd.options.display.float_format = '{:,.2f}'.format

import warnings
warnings.simplefilter("ignore", category=FutureWarning)
````

````python
loan = pd.read_pickle('data_input/loan_clean')
loan.head()
````

## Overview

Di bagian Overview, kita akan menampilkan metrik utama untuk memberikan gambaran singkat tentang data pinjaman. Berikut ini metrik yang akan kita hitung dan tampilkan:

1. **Total Loans**: Metrik ini menunjukkan jumlah total pinjaman dalam kumpulan data.
2. **Total Loan Amount**: Metrik ini memberikan jumlah semua jumlah pinjaman dalam kumpulan data.
3. **Average Interest Rate**: Metrik ini menghitung rata-rata suku bunga di semua pinjaman.
4. **Average Loan Amount**: Metrik ini menunjukkan rata-rata jumlah pinjaman.

ðŸ’¡ **Hint**:

1. Gunakan format string dengan syntax `f"Harga = Rp {variable_price}"` untuk menampilkan isi variabel didalam string.
2. Tambahkan `:,.0f` dibelakang variabel berisi angka (`f"Harga = Rp {variable_price:,.0f}"`) untuk memberikan pemisah ribuan menggunakan koma dan membulatkan desimal ke satuan terdekat.

````python
# code here
print("Total Loans")
f"{loan['id'].count():,.0f}"
#f"{loan['id'].count():,.0f}".replace(',','.')
````

````python
# code here
print("Total Loan Amount")
f"${loan['loan_amount'].sum():,.0f}"
````

````python
# code here
print("Average Interest Rate")
f"{loan['interest_rate'].mean():,.0f}"
````

````python
# code here
print("Average Loan Amount")
f"{loan['loan_amount'].mean():,.0f}"
````

## Time-Based Analysis

Di bagian ini, kita akan menganalisis tren data pinjaman dari waktu ke waktu. Hal tersebut termasuk memeriksa jumlah pinjaman yang diberikan, jumlah total pinjaman dari waktu ke waktu, dan distribusi pinjaman berdasarkan hari dalam seminggu.

**1. Loans Issued Over Time**

Kita akan membuat grafik yang menunjukkan jumlah pinjaman yang diterbitkan dari waktu ke waktu. Ini akan membantu mengidentifikasi tren, seperti periode dengan penerbitan pinjaman tinggi atau rendah.

````python
loan_date_count = loan.groupby('issue_date')['loan_amount'].count()
loan_date_count
````

**2. Loan Amount Over Time**

Kita akan membuat grafik untuk menampilkan jumlah total pinjaman yang dikeluarkan dari waktu ke waktu. Visualisasi ini akan membantu memahami tren dalam jumlah total pinjaman dan mengidentifikasi perubahan signifikan selama periode yang berbeda.

````python
loan_date_sum = loan.groupby('issue_date')['loan_amount'].sum()
loan_date_sum
````

**3. Issue Date Analysis**

Kita akan membuat diagram untuk menunjukkan distribusi pinjaman berdasarkan hari dalam seminggu. Ini akan membantu menganalisis apakah ada hari-hari tertentu saat pinjaman lebih sering diterbitkan.

````python
loan_day_count = loan.groupby('issue_weekday')['loan_amount'].count()
loan_day_count
````

## Loan Performance

Pada bagian ini, kita akan menganalisis dan memvisualisasikan kinerja pinjaman berdasarkan berbagai kriteria. Bagian ini akan mencakup:

**1. Loan Condition Analysis**

Kita akan menampilkan distribusi pinjaman berdasarkan kondisinya menggunakan **pie chart**. Visualisasi ini akan membantu memahami proporsi pinjaman dalam berbagai kondisi.

````python
# code here
px.pie(
    loan,
    names='loan_condition',
    hole=0.4,
    title='Distribution of Conditions',
)
````

**2. Grade Distribution**

Selanjutnya, kita akan membuat diagram untuk menunjukkan distribusi pinjaman berdasarkan nilainya. Visualisasi ini memberikan wawasan tentang bagaimana pinjaman dikategorikan berdasarkan nilai dan frekuensinya.

````python
grade = loan['grade'].value_counts().sort_index()

print(grade)
````

````python
# code here
px.bar(
    grade,
    title='Distribution of Loans by Grade',
    labels={
        'value':'Numbers of Loans'
    }
).update_layout(showlegend=False)
````

## Financial Analysis

Pada bagian ini, kita akan fokus pada analisis aspek keuangan pinjaman yang diklasifikasikan sebagai `"Good Loan"` dan `"Bad Loan"`. Proses ini akan membantu kita memahami distribusi jumlah pinjaman dan variasinya di berbagai tujuan.

````python
loan_condition = loan[loan['loan_condition'] == 'Good Loan']
````

**1. Loan Amount Distribution**

Kita akan membuat **Histogram** untuk menampilkan distribusi jumlah pinjaman untuk pinjaman yang diklasifikasikan sebagai `"Good Loan"` dan memisahkannya berdasarkan berapa lama `term` peminjamannya. Visualisasi ini akan membantu kami memahami rentang dan frekuensi jumlah pinjaman dalam kategori ini.

<details>
  <summary>Double-click this markdown to show additional enhancement features</summary>
  
```python
    template='seaborn',
    labels={
        'loan_amount':'Loan Amount',
        'term':'Loan Term'
	}
```

</details>

````python
# code here
px.histogram(
    loan_condition,
    x='loan_amount',
    color='term',
    nbins=20,
    template='seaborn',
    labels={
        'loan_amount':'Loan Amount',
        'term':'Loan Term'
    }
)
````

**2. Loan Amount Distribution by Purpose**

Kita akan membuat **Boxplot** untuk menunjukkan distribusi jumlah pinjaman untuk setiap `purpose` di antara pinjaman yang diklasifikasikan sebagai `"Good Loan"` dan memisahkannya berdasarkan berapa lama `term` peminjamannya. Ini akan membantu kita melihat bagaimana jumlah pinjaman bervariasi di berbagai tujuan dan apakah ada perbedaan yang signifikan.

<details>
  <summary>Double-click this markdown to show additional enhancement features</summary>

	title="Loan Amount Distribution by Purpose",

````python
# code here
px.box(
    loan_condition,
    x='purpose',
    y='loan_amount',
    color='term',
    template='seaborn',
    labels={
        'term':'Loan Term',
        'loan_amount':'Loan Amount',
        
    }
)
````
